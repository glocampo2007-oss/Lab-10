```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

### 2. Cargar los datos

```python
# Cargar el dataset Iris
iris = load_iris()

# Extraer solo las dos primeras características (longitud y ancho del sépalo)
X = iris.data[:, :2]  # Todas las filas, columnas 0 y 1

# Extraer las etiquetas de la especie
y = iris.target

print(f"Dimensiones de las características (X): {X.shape}")
print(f"Dimensiones de las etiquetas (y): {y.shape}")
print("Primeras 5 filas de características (X):")
print(X[:5])
print("Primeras 5 etiquetas (y):")
print(y[:5])
```

### 3. Dividir el conjunto de datos

```python
# Dividir el conjunto de datos en entrenamiento (70%) y prueba (30%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(f"Tamaño del conjunto de entrenamiento de características (X_train): {X_train.shape}")
print(f"Tamaño del conjunto de prueba de características (X_test): {X_test.shape}")
print(f"Tamaño del conjunto de entrenamiento de etiquetas (y_train): {y_train.shape}")
print(f"Tamaño del conjunto de prueba de etiquetas (y_test): {y_test.shape}")
```

### 4. Crear el modelo de clasificación

```python
# Crear el modelo de regresión logística
model = LogisticRegression()

print("Modelo de Regresión Logística creado:")
print(model)
```

### 5. Entrenar el modelo

```python
# Entrenar el modelo con los datos de entrenamiento
model.fit(X_train, y_train)

print("Modelo entrenado exitosamente.")
```

### 6. Hacer predicciones

```python
# Hacer predicciones sobre el conjunto de prueba
y_pred = model.predict(X_test)

print("Primeras 5 predicciones:")
print(y_pred[:5])
print("Primeras 5 etiquetas reales del conjunto de prueba:")
print(y_test[:5])
```

### 7. Calcular la precisión

```python
# Calcular la precisión del modelo
accuracy = accuracy_score(y_test, y_pred)

print(f"Precisión del modelo de Regresión Logística: {accuracy:.4f}")
```

### 8. Visualización de los resultados

```python
# Crear un DataFrame para facilitar la visualización
# Usamos el conjunto de prueba para la visualización de predicciones
df_results = pd.DataFrame({
    'sepal length (cm)': X_test[:, 0],
    'sepal width (cm)': X_test[:, 1],
    'Actual Species': y_test,
    'Predicted Species': y_pred
})

# Mapear los números de las especies a sus nombres originales
species_names = iris.target_names
df_results['Actual Species Name'] = df_results['Actual Species'].map(lambda x: species_names[x])
df_results['Predicted Species Name'] = df_results['Predicted Species'].map(lambda x: species_names[x])

plt.figure(figsize=(15, 7))

# Gráfico de dispersión para las especies reales
plt.subplot(1, 2, 1)
sns.scatterplot(
    x='sepal length (cm)',
    y='sepal width (cm)',
    hue='Actual Species Name',
    data=df_results,
    palette='viridis',
    s=100,
    alpha=0.8
)
plt.title('Especies Reales (Conjunto de Prueba)')
plt.xlabel('Longitud del Sépalo (cm)')
plt.ylabel('Ancho del Sépalo (cm)')
plt.grid(True)

# Gráfico de dispersión para las especies predichas
plt.subplot(1, 2, 2)
sns.scatterplot(
    x='sepal length (cm)',
    y='sepal width (cm)',
    hue='Predicted Species Name',
    data=df_results,
    palette='viridis',
    s=100,
    alpha=0.8
)
plt.title('Especies Predichas (Conjunto de Prueba)')
plt.xlabel('Longitud del Sépalo (cm)')
plt.ylabel('Ancho del Sépalo (cm)')
plt.grid(True)

plt.tight_layout()
plt.show()
```
